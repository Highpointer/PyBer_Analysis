```python
# Combine the data into a single dataset. Add the following code to a new cell and run the cell to merge the two DataFrames. 5.4.2
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the merged DataFrame
pyber_data_df.head(149)
```

```python
# Create the Suburban and Rural city DataFrames. 5.3.1
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_cities_df
```

```python
# Get the number of rides for urban cities. 5.3.2
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
urban_ride_count.head(10)
```

```python
# Create the suburban and rural ride count. 5.3.2
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
```

```python
# Get average fare for each city in the urban cities 5.3.3
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
urban_avg_fare.head()
```

```python
# Get average fare for each city in the suburban and rural cities. 5.3.3
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
# Now we have two of three datasets we need to create a bubble chart.
```

```python
# To get the average number of drivers for each city in the urban, suburban, and rural cities,
# we can use the groupby() function and get the mean() of the driver_count column. 5.3.4
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
urban_driver_count.head(10)
```

```python
# Get the average number of drivers for each city for the suburban and rural cities. 5.3.4
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
```

```python
# Build the scatter plots for urban cities. 5.3.5 Create Bubble Charts
plt.scatter(urban_ride_count, urban_avg_fare)
```

```python
# Build the scatter plots for urban cities. 5.3.5
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=urban_driver_count)
```

```python
# Some of the bubbles are still too small, so let's increase the size by a factor of 10 and add a black edge color to the
# circles that have a line width of 1. While we're changing these features, let's pay homage to the company's color scheme
# and make the urban markers coral and 20% transparent. We can also add a title, labels for the axes, a legend, and a grid.

# Build the scatter plots for urban cities. 5.3.5
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Build the scatter plots for suburban cities. Change the colour to sky blue. 5.3.5
plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Build the scatter plots for rural cities. 5.3.5
# Change the x- and y-axes variables and the size parameter to the rural city data and change the color of the circle to gold.
plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Build the scatter charts for each city type. 5.3.6
plt.subplots(figsize=(10, 6))
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Incorporate a text label about circle size.
plt.text(42, 35, "Note: Circle size correlates", fontsize="12")
plt.text(42, 33.5, "with driver count per city", fontsize = "12")
# Save the figure.
plt.savefig("Resources/Fig1.png")

# Show the plot
plt.show()
```
